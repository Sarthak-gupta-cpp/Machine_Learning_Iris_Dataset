# Making The Model

## Imports for the model

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
```

## Train-test split


```python
df_train, df_test = train_test_split(df, test_size=0.25)
```

```python
df_test.head()
```

## Prepare our data for training
-> Convert to numpy arrays

-> sklearn accepts numpy arrays

```python
x_train = df_train.drop(columns=["target", "target_name"]).values
y_train = df_train["target"].values

x_test = df_test.drop(columns=["target", "target_name"]).values
y_test = df_test["target"].values
```

```python
x_train.shape
```

## Modelling
-> Our baseline is 33% (random guessing)

-> Any model should have accuracy greater than 33%

-> Will implement Logisitic rregression

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# make a validation set
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
model2 = LogisticRegression(max_iter=1000)
model2.fit(xt, yt)
```

```python
model2.score(xv, yv)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
#This makes different validation sets on its own
accuracies = cross_val_score(model2, x_train, y_train, cv=5, scoring="accuracy")
```

```python
np.mean(accuracies)
```

## Can we do better??
-> Will train another model on random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model3 = RandomForestClassifier()
```

```python
accs = cross_val_score(model3, x_train, y_train, cv=5, scoring="accuracy")
np.mean(accs)
```

## Model Tuning
-> changing the hyperparameters to maximise

```python
model4 = LogisticRegression(max_iter=300, C=1)
accs = cross_val_score(model4, x_train, y_train, cv=5, scoring="accuracy")
np.mean(accs)

```

## Test on Test Data

```python
model4.fit(x_train, y_train)
```

```python
model4.score(x_test, y_test)
```

```python
y_test_pred = model4.predict(x_test)
correct = y_test_pred == y_test
np.mean(correct)
```
